```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Code to create a black image (use any one code)

```python
# imgA = np.zeros((1024,1024,3),dtype=np.uint8)
# plt.imshow(imgA)

imgA = np.zeros((1024,1024),dtype=np.uint8)
plt.imshow(imgA,cmap=plt.cm.bone)
plt.title('Image a (Black)')
```

Code to create a white image

```python
imgB = np.ones((1024,1024,3),dtype=np.uint8)
imgB = imgB*255
plt.imshow(imgB)
plt.title('Image b (White)')
```

Creating image with horizontal white lines on black image

```python
imgC = np.zeros((1024, 1024, 3), dtype=np.uint8)
for i in range(0, 1024, 128):
  if (i // 128) % 2 == 0:
    imgC[i:i+128, :] = 255

plt.imshow(imgC)
plt.title('Image c (Horizontal)')
```

Creating image with vertical white lines on black image

```python
imgD = np.zeros((1024, 1024, 3), dtype=np.uint8)
for i in range(0, 1024, 128):
    if (i // 128) % 2 == 0:
        imgD[:, i:i+128] = 255
  
plt.imshow(imgD)
plt.title('Image d (Vertical)')
```

d. Create a grayscale image of size 256x1024. Intensity of image 
should vary sinusoidal.

```python
height, width = 256, 1024

image = np.zeros((height, width), dtype=np.uint8)

frequency = 2 * np.pi / width

for y in range(height):
    for x in range(width):
        intensity = 127.5 * (1 + np.sin(frequency * x))
        image[y, x] = intensity

plt.imshow(image, cmap='gray')
plt.title('Grayscale Image with Sinusoidal Intensity Variation')
```
